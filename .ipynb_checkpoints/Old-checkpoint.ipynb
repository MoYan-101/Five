```python
import os
import re
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.decomposition import NMF
import matplotlib.pyplot as plt
import seaborn as sns

# Â∞ùËØïÂØºÂÖ•Ëß£ÊûêÂ∫ìÂíåIPythonÊòæÁ§∫ÂäüËÉΩ
try:
    import bibtexparser
    import rispy
    from IPython.display import display
except ImportError as e:
    print(f"ÈîôËØØ: Áº∫Â∞ëÂøÖË¶ÅÁöÑÂ∫ì„ÄÇËØ∑Âú®ÂçïÂÖÉÊ†º‰∏≠ËøêË°å: !pip install bibtexparser rispy pandas scikit-learn matplotlib seaborn")
    # ‰∏∫‰∫ÜÈò≤Ê≠¢ÂÆåÂÖ®Â¥©Ê∫ÉÔºåÂÆö‰πâ‰∏Ä‰∏™ÂÅáÁöÑ display
    def display(x): print(x)

# ËÆæÁΩÆÁªòÂõæÈ£éÊ†º
sns.set_style("whitegrid")
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS'] # Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫‰∏≠ÊñáÊ†áÁ≠æ
plt.rcParams['axes.unicode_minus'] = False

# === Êñ∞Â¢ûÔºöËá™ÂÆö‰πâÂ≠¶ÊúØÂÅúÁî®ËØçË°® ===
# Ëøô‰∫õËØçÂú®ÊëòË¶Å‰∏≠Âá∫Áé∞È¢ëÁéáÊûÅÈ´òÔºå‰ΩÜÈÄöÂ∏∏‰∏ç‰ª£Ë°®ÂÖ∑‰ΩìÁöÑÁ†îÁ©∂‰∏ªÈ¢ò
ACADEMIC_STOP_WORDS = list(ENGLISH_STOP_WORDS) + [
    'study', 'paper', 'article', 'research', 'review',
    'result', 'results', 'finding', 'findings', 'conclusion', 'conclusions',
    'method', 'methods', 'methodology', 'approach', 'approaches', 'algorithm',
    'model', 'models', 'modeling', 'framework', 'system',
    'use', 'used', 'using', 'utilize', 'utilized', 'based', 'propose', 'proposed',
    'aim', 'aims', 'goal', 'objective', 'purpose',
    'analysis', 'experiment', 'experimental', 'data', 'dataset',
    'high', 'low', 'performance', 'significant', 'significantly',
    'demonstrate', 'show', 'shown', 'present', 'presents', 'provide', 'provides',
    'introduction', 'background', 'discuss', 'discussion',
    'compare', 'comparison', 'evaluation', 'validate', 'validation',
    'time', 'year', 'new', 'novel', 'different', 'various','government','governments'
]

class LiteratureAnalyzer:
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = []
        self.df = None
        self.feature_names = []
        self.tfidf_matrix = None
        self.nmf_model = None

    def load_data(self):
        """Âä†ËΩΩÊï∞ÊçÆÂπ∂ËΩ¨Êç¢‰∏∫ DataFrame"""
        if not self.filepath or not os.path.exists(self.filepath):
            print(f"‚ùå ÈîôËØØÔºöÊâæ‰∏çÂà∞Êñá‰ª∂ '{self.filepath}'")
            print("ËØ∑Âú®‰ª£Á†ÅÂ∫ïÈÉ®‰øÆÊîπ target_filename ‰∏∫‰Ω†ÁöÑ .bib Êàñ .ris Êñá‰ª∂Ë∑ØÂæÑ„ÄÇ")
            return False

        ext = os.path.splitext(self.filepath)[1].lower()
        print(f"üìÇ Ê≠£Âú®ËØªÂèñÊñá‰ª∂: {self.filepath} ...")
        
        try:
            if ext == '.ris':
                with open(self.filepath, 'r', encoding='utf-8', errors='ignore') as f:
                    entries = rispy.load(f)
                    for entry in entries:
                        self.data.append({
                            'title': entry.get('primary_title') or entry.get('title', 'No Title'),
                            'abstract': entry.get('abstract') or entry.get('notes', ''),
                            'year': entry.get('year', '')
                        })
            elif ext == '.bib':
                with open(self.filepath, 'r', encoding='utf-8', errors='ignore') as f:
                    bib_database = bibtexparser.load(f)
                    for entry in bib_database.entries:
                        self.data.append({
                            'title': entry.get('title', 'No Title'),
                            'abstract': entry.get('abstract', ''),
                            'year': entry.get('year', '')
                        })
            else:
                print("‚ùå ‰∏çÊîØÊåÅÁöÑÊñá‰ª∂Ê†ºÂºè„ÄÇ‰ªÖÊîØÊåÅ .ris Êàñ .bib")
                return False
            
            self.df = pd.DataFrame(self.data)
            # ËøáÊª§ÊéâÊëòË¶ÅÂ§™Áü≠ÁöÑ
            self.df = self.df[self.df['abstract'].str.len() > 20].copy()
            print(f"‚úÖ ÊàêÂäüÂä†ËΩΩ {len(self.df)} Êù°ÊúâÊïàÊñáÁåÆ")
            # Âú® Jupyter ‰∏≠Áõ¥Êé•Â±ïÁ§∫ÂâçÂá†Ë°åÊï∞ÊçÆ
            print("Êï∞ÊçÆÈ¢ÑËßà:")
            display(self.df[['title', 'year']].head())
            return True
            
        except Exception as e:
            print(f"‚ùå ËØªÂèñÂ§±Ë¥•: {e}")
            return False

    def analyze_keywords(self, top_n=15, plot=True):
        """ÊèêÂèñÂπ∂ÂèØËßÜÂåñÂÖ≥ÈîÆËØç"""
        print("\nüîç Ê≠£Âú®ÂàÜÊûêÊ†∏ÂøÉÂÖ≥ÈîÆËØç (Â∑≤ËøáÊª§ÈÄöÁî®Â≠¶ÊúØËØçÊ±á)...")
        
        # ‰ΩøÁî®Ëá™ÂÆö‰πâÂÅúÁî®ËØçÔºåÂπ∂Ë∞ÉÊï¥ max_df
        # max_df=0.80 ÊÑèÂë≥ÁùÄÂ¶ÇÊûú‰∏Ä‰∏™ËØçÂú®Ë∂ÖËøá 80% ÁöÑÊñáÊ°£‰∏≠Âá∫Áé∞ÔºåÂ∞±ÂøΩÁï•ÂÆÉ
        vectorizer = TfidfVectorizer(
            stop_words=ACADEMIC_STOP_WORDS, 
            max_features=1000, 
            max_df=0.80, 
            min_df=2
        )
        
        try:
            self.tfidf_matrix = vectorizer.fit_transform(self.df['abstract'])
            self.feature_names = vectorizer.get_feature_names_out()
            
            mean_tfidf = np.array(self.tfidf_matrix.mean(axis=0)).flatten()
            top_indices = mean_tfidf.argsort()[-top_n:][::-1]
            
            keywords = [self.feature_names[i] for i in top_indices]
            scores = [mean_tfidf[i] for i in top_indices]
            
            if plot:
                plt.figure(figsize=(10, 6))
                # ‰øÆÂ§ç FutureWarning: Assign y variable to hue
                sns.barplot(x=scores, y=keywords, hue=keywords, palette="viridis", legend=False)
                plt.title(f"Top {top_n} Ê†∏ÂøÉÂÖ≥ÈîÆËØç (ÂéªÂô™Âêé)")
                plt.xlabel("ÈáçË¶ÅÊÄßÂàÜÊï∞ (TF-IDF)")
                plt.show()
                
            return keywords
        except ValueError:
            print("‚ö†Ô∏è Êï∞ÊçÆÈáèÂ§™Â∞ëÊàñËØçÊ±áËøá‰∫éÁ®ÄÁñèÔºåÊó†Ê≥ïËøõË°åÊúâÊïàÂàÜÊûê„ÄÇ")
            return []

    def extract_topics(self, n_topics=3, plot=True):
        """ÊèêÂèñÊΩúÂú®‰∏ªÈ¢òÂπ∂ÂèØËßÜÂåñ"""
        print(f"\nüß† Ê≠£Âú®ÊèêÂèñ {n_topics} ‰∏™ÊΩúÂú®Á†îÁ©∂Áª¥Â∫¶ (Topic Modeling)...")
        
        if self.tfidf_matrix is None:
            return

        self.nmf_model = NMF(n_components=n_topics, random_state=42, init='nndsvd')
        topic_values = self.nmf_model.fit_transform(self.tfidf_matrix)
        
        # ‰øùÂ≠ò‰∏ªÈ¢òÂàÜÁ±ªÁªìÊûú
        self.df['Topic_ID'] = topic_values.argmax(axis=1) + 1
        
        # ÊâìÂç∞ÊØè‰∏™‰∏ªÈ¢òÁöÑÂÖ≥ÈîÆËØç
        print("-" * 50)
        for topic_idx, topic in enumerate(self.nmf_model.components_):
            top_features_ind = topic.argsort()[:-8:-1]
            top_words = [self.feature_names[i] for i in top_features_ind]
            topic_name = f"Áª¥Â∫¶ {topic_idx + 1}"
            print(f"üìå {topic_name}: {', '.join(top_words)}")
        print("-" * 50)

        # ÂèØËßÜÂåñÔºöÊñáÊ°£Âú®ÂêÑ‰∏™Áª¥Â∫¶ÁöÑÂàÜÂ∏ÉÊÉÖÂÜµ
        if plot:
            plt.figure(figsize=(8, 5))
            # ‰øÆÂ§ç FutureWarning: Assign x variable to hue
            sns.countplot(x='Topic_ID', data=self.df, hue='Topic_ID', palette="pastel", legend=False)
            plt.title("ÊñáÁåÆ‰∏ªÈ¢òÂàÜÂ∏ÉÁªüËÆ°")
            plt.xlabel("ÊΩúÂú®Áª¥Â∫¶ (Topic ID)")
            plt.ylabel("ÊñáÁåÆÊï∞Èáè")
            plt.show()

    def extract_key_sentences(self):
        """ÊèêÂèñÊëòË¶Å‰∏≠ÁöÑÁªìÊûÑÂåñÂè•Â≠êÔºàÁõÆÁöÑ„ÄÅÊñπÊ≥ï„ÄÅÁªìÊûúÔºâ"""
        print("\nüìù Ê≠£Âú®‰ªéÊëòË¶Å‰∏≠ÊèêÂèñÂÖ≥ÈîÆÂè•...")
        
        patterns = {
            'Objective (ÁõÆÁöÑ)': [r'aim to', r'purpose', r'goal', r'investigate', r'focus'],
            'Method (ÊñπÊ≥ï)': [r'method', r'propose', r'utilize', r'experiment', 'network', 'via', 'through'],
            'Result (ÁªìÊûú)': [r'show', r'found', r'demonstrate', r'achieve', 'outperform', 'improve']
        }
        
        results = []
        for text in self.df['abstract']:
            row_res = {'Objective': '', 'Method': '', 'Result': ''}
            # ÁÆÄÂçïÁöÑÂàÜÂè•
            sentences = re.split(r'[.!?]\s+', text)
            for sent in sentences:
                sent_lower = sent.lower()
                for cat, keys in patterns.items():
                    key_name = cat.split()[0]
                    # Âè™ÊúâÂΩìËØ•Â≠óÊÆµ‰∏∫Á©∫Êó∂ÊâçÂ°´ÂÖ•ÔºàÂèñÂá∫Áé∞ÁöÑÁ¨¨‰∏Ä‰∏™Áõ∏ÂÖ≥Âè•Â≠êÔºâ
                    if not row_res[key_name] and any(k in sent_lower for k in keys):
                        row_res[key_name] = sent
            results.append(row_res)
            
        # Â∞ÜÁªìÊûúÂêàÂπ∂Âõû DataFrame
        structure_df = pd.DataFrame(results)
        # ÈÅøÂÖçÂàóÂêçÈáçÂ§ç
        self.df = self.df.loc[:, ~self.df.columns.isin(['Objective', 'Method', 'Result'])]
        self.df = pd.concat([self.df, structure_df], axis=1)
        
        print("ÊäΩÊ†∑Â±ïÁ§∫ÊèêÂèñÁªìÊûú:")
        display(self.df[['title', 'Objective', 'Method', 'Result']].head(3))

    def save_results(self):
        output_file = "analysis_report.csv"
        self.df.to_csv(output_file, index=False, encoding='utf_8_sig')
        print(f"\nüíæ ÂÆåÊï¥ÁªìÊûúÂ∑≤‰øùÂ≠òËá≥: {output_file}")

# ==========================================
# üëá ËØ∑Âú®ËøôÈáå‰øÆÊîπÊñá‰ª∂Âêç
# ==========================================

# 1. ËÆæÁΩÆ‰Ω†ÁöÑÊñá‰ª∂Âêç (ÂøÖÈ°ª‰øÆÊîπËøôÈáåÔºÅ‰æãÂ¶Ç: "my_references.bib")
target_filename = "ktss.bib"  

# 2. ÊÉ≥Ë¶ÅÊèêÂèñÂá†‰∏™‰∏ªÈ¢ò/Áª¥Â∫¶Ôºü
num_topics = 10

# --- ÊâßË°åÈÄªËæë ---
analyzer = LiteratureAnalyzer(target_filename)

if analyzer.load_data():
    # 1. ÂàÜÊûêÂπ∂ÁîªÂá∫ÂÖ≥ÈîÆËØç
    analyzer.analyze_keywords(top_n=15, plot=True)
    
    # 2. ÊèêÂèñÊΩúÂú®Áª¥Â∫¶Âπ∂ÁîªÂá∫ÂàÜÂ∏É
    analyzer.extract_topics(n_topics=num_topics, plot=True)
    
    # 3. ÊèêÂèñÊëòË¶ÅÁªìÊûÑ (ÁõÆÁöÑ/ÊñπÊ≥ï/ÁªìÊûú)
    analyzer.extract_key_sentences()
    
    # 4. ‰øùÂ≠ò
    analyzer.save_results()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. ÂÆö‰πâÊò†Â∞ÑÂÖ≥Á≥ª (Ê†πÊçÆÂàöÊâçÁöÑÂàÜÊûê)
# Ê≥®ÊÑèÔºöPythonÁ¥¢ÂºïÊòØ‰ªé0ÂºÄÂßãÁöÑÔºåÊâÄ‰ª•Áª¥Â∫¶1ÊòØTopic 0
topic_mapping = {
    1: "Operating Logic & Process",  # ÂéüÁª¥Â∫¶ 1
    2: "Leadership",                 # ÂéüÁª¥Â∫¶ 2
    3: "Tech-Driven Logic",          # ÂéüÁª¥Â∫¶ 3 (‰Ωú‰∏∫ Process ÁöÑ‰∏ÄÈÉ®ÂàÜ)
    4: "Admin Power & Bureaucracy",  # ÂéüÁª¥Â∫¶ 4
    5: "Public Service Delivery",    # ÂéüÁª¥Â∫¶ 5
    6: "Org Structure & Collab"      # ÂéüÁª¥Â∫¶ 6 (ÂéüÁª¥Â∫¶ 7,8 ÊöÇÊó∂ÂøΩÁï•ÊàñÂêàÂπ∂)
}

# 2. Â∞Ü DataFrame ‰∏≠ÁöÑÊï∞Â≠ó ID ÊõøÊç¢‰∏∫‰Ω†ÁöÑÁêÜËÆ∫ÂêçÁß∞
# ÂÅáËÆæ analyzer.df ËøòÂú®ÂÜÖÂ≠ò‰∏≠
if analyzer.df is not None and 'Topic_ID' in analyzer.df.columns:
    # ÂàõÂª∫‰∏Ä‰∏™Êñ∞ÁöÑÂàóÁî®‰∫éÁîªÂõæ
    analyzer.df['Dimension_Name'] = analyzer.df['Topic_ID'].map(topic_mapping)
    
    # ËøáÊª§ÊéâÊ≤°ÊúâÊò†Â∞ÑÂà∞ÁöÑÁª¥Â∫¶ÔºàÂ¶ÇÊûú‰Ω†ÊèêÂèñ‰∫Ü8‰∏™‰ΩÜÂè™Áî®‰∫Ü6‰∏™Ôºâ
    plot_df = analyzer.df.dropna(subset=['Dimension_Name'])

    # 3. ÁªòÂà∂ÊúÄÁªàÁöÑÂØπÂ∫îÂõæ
    plt.figure(figsize=(10, 6))
    sns.countplot(y='Dimension_Name', data=plot_df, 
                  order=plot_df['Dimension_Name'].value_counts().index,
                  palette="viridis", hue='Dimension_Name', legend=False)
    
    plt.title("ÊñáÁåÆÂú® '‰∫îÂ§ßÂÖ≥ÈîÆÁª¥Â∫¶' ‰∏äÁöÑÂàÜÂ∏É (NMFËÅöÁ±ªÊò†Â∞Ñ)", fontsize=14)
    plt.xlabel("ÊñáÁåÆÊï∞Èáè", fontsize=12)
    plt.ylabel("‰Ω†ÁöÑÁêÜËÆ∫Áª¥Â∫¶", fontsize=12)
    plt.show()
else:
    print("ËØ∑ÂÖàËøêË°å‰πãÂâçÁöÑÂàÜÊûê‰ª£Á†Å‰ª•ÁîüÊàê analyzer.df")
```
